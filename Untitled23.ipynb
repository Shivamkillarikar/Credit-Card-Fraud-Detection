```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"C:\Users\Srushti\OneDrive\Desktop\creditcard.csv")
```

```python
df.head()
```

```python
#Dataset Information
df.info()
```

```python
#checking for null value
df.isnull().sum()
```

```python
df.describe()
```

```python
#Count of legit transactions & fake transactions
df['Class'].value_counts()
```

```python
# 0->Normal Transaction
# 1->Fraudelent Transaction
#Imbalanced Data
```

```python
authentic=df[df.Class==0]
fake=df[df.Class==1]
```

```python
print(authentic.shape)
print(fake.shape)
```

```python
authentics=authentic.sample(n=492)
```

```python
print(authentics.shape)
```

```python
new_df = pd.concat([authentics,fake],axis=0)
```

```python
new_df.head()
```

```python
#Dataset Is Balanced Now
new_df['Class'].value_counts()
```

```python
#Splitting the data into input values(Features) and targets
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training data and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training, Logistic Regression
model = LogisticRegression()
```

```python
#Training the model with the training data
model.fit(X_train,Y_train)
```

```python
#Model evaluation
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data : " , training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on training data : " , test_data_accuracy)
```
